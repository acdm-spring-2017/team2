```python
import plotly.plotly as py
import plotly.graph_objs as go

Women = go.Bar(
    x= neighborhoodW,
    y= ageW,
    name='Women',
    text= offensesW,
    marker = dict( color='#F5B7B1')
)
Men = go.Bar(
    x= neighborhoodM,
    y= ageM,
    name='Men',
    text= offensesM,
    marker = dict( color = '#A9CCE3')
)
WhiteWomen = go.Bar(
    x = neighborhoodWW,
    y= ageWW,
    name='White Women',
    text= offensesWW,
    marker = dict( color ='#F5B041')
)
WhiteMen = go.Bar(
    x= neighborhoodWM,
    y= ageWM,
    name='White Men',
    text= offensesWM,
    marker = dict( color = '#F5B041')
)


BlackWomen = go.Bar(
    x = neighborhoodBW,
    y= ageBW,
    name='Black Women',
    text= offensesBW,
    marker = dict( color ='#E67E22')
)

BlackMen = go.Bar(
    x = neighborhoodBM,
    y= ageBM,
    name='Black Men',
    text= offensesBM,
    marker = dict( color ='#E67E22')
)

AsianWomen = go.Bar(
    x = neighborhoodAW,
    y= ageAW,
    name='Asian Women',
    text= offensesAW,
    marker = dict( color ='#5D6D7E')
)

AsianMen = go.Bar(
    x = neighborhoodAM,
    y= ageAM,
    name='Asian Men',
    text= offensesAM,
    marker = dict( color ='#5D6D7E')
)


HispanicWomen = go.Bar(
    x = neighborhoodHW,
    y= ageHW,
    name='Hispanic Women',
    text= offensesHW,
    marker = dict( color ='#F1C40F')
)

HispanicMen = go.Bar(
    x = neighborhoodHM,
    y= ageHM,
    name='Hispanic Men',
    text= offensesHM,
    marker = dict( color ='#F1C40F')
)

OtherWomen = go.Bar(
    x = neighborhoodOW,
    y= ageOW,
    name='Other Women',
    text= offensesOW,
    marker = dict( color ='#BDC3C7')
)

OtherMen = go.Bar(
    x = neighborhoodOM,
    y = ageOM,
    name='Other Men',
    text= offensesOM,
    marker = dict( color ='#BDC3C7')
)




data = [Women, Men, WhiteWomen, WhiteMen, BlackWomen, BlackMen, AsianWomen, AsianMen, HispanicWomen, 
        HispanicMen, OtherWomen, OtherMen]
layout = go.Layout(
    yaxis=dict(
        title='Age',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#17202A'
        )
    ),
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)',
    barmode='stack',
    title='Arrests in Allegheny County by Neighborhood',
    updatemenus=list([
        dict(
            x=-0.05,
            y=1,
            yanchor='top',
            buttons=list([
                dict(
                    args=['visible', [True, True, True, True, True, True, False, False, False, False, False, False]],
                    label='All',
                    method='restyle'
                ),
                dict(
                    args=['visible', [True, True, False, False, False, False, False, False, False, False, False, False]],
                    label='Men and Women',
                    method='restyle'
                ),
                dict(
                    args=['visible', [False, False, True, False, True, False, True, False, True, False, True, False]],
                    label='Women by Race',
                    method='restyle'
                ),
                dict(
                    args=['visible', [False, False, False, True, False, True, False, True, False, True, False, True]],
                    label='Men by Race',
                    method='restyle'
                ),
                dict(
                    args=['visible', [False, False, False, False, True, False, True, False, True, False, True, False]],
                    label='Women of Color',
                    method='restyle'
                ),
                 dict(
                    args=['visible', [False, False, False, False, False, True, False, True, False, True, False, True]],
                    label='Men of Color',
                    method='restyle'
                )
            ])),
    ]),
)


fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='stacked-bar')
```
