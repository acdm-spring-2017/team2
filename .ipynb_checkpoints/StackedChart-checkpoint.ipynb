```python
import pandas as pd
df2 =  pd.read_csv('arrestDataW.csv')
from pandas import *

raceW = list(df2["RACE"])
ageW = list(df2["AGE"])
genderW = list(df2["GENDER"])
arresttimeW = list(df2["ARRESTTIME"])
arrestlocationW = list(df2["ARRESTLOCATION"])
offensesW = list(df2["OFFENSES"])
locationW = list(df2["INCIDENTLOCATION"])
neighborhoodW = list(df2["INCIDENTNEIGHBORHOOD"])
zoneW = list(df2["INCIDENTZONE"])
latW = list(df2["X"])
longW = list(df2["Y"])

import pandas as pd
df =  pd.read_csv('arrestDataM.csv')
from pandas import *

raceM = list(df["RACE"])
ageM = list(df["AGE"])
genderM = list(df["GENDER"])
arresttimeM = list(df["ARRESTTIME"])
arrestlocationM = list(df["ARRESTLOCATION"])
offensesM = list(df["OFFENSES"])
locationM = list(df["INCIDENTLOCATION"])
neighborhoodM = list(df["INCIDENTNEIGHBORHOOD"])
zoneM = list(df["INCIDENTZONE"])
latM = list(df["X"])
longM = list(df["Y"])

import pandas as pd
df =  pd.read_csv('arrestDataWhiteW.csv')
from pandas import *

raceWW = list(df["RACE"])
ageWW = list(df["AGE"])
genderWW = list(df["GENDER"])
arresttimeWW = list(df["ARRESTTIME"])
arrestlocationWW = list(df["ARRESTLOCATION"])
offensesWW = list(df["OFFENSES"])
locationWW = list(df["INCIDENTLOCATION"])
neighborhoodWW = list(df["INCIDENTNEIGHBORHOOD"])
zoneWW = list(df["INCIDENTZONE"])
latWW = list(df["X"])
longWW = list(df["Y"])

import pandas as pd
df =  pd.read_csv('arrestDataWhiteM.csv')
from pandas import *

raceWM = list(df["RACE"])
ageWM = list(df["AGE"])
genderWM = list(df["GENDER"])
arresttimeWM = list(df["ARRESTTIME"])
arrestlocationWM = list(df["ARRESTLOCATION"])
offensesWM = list(df["OFFENSES"])
locationWM = list(df["INCIDENTLOCATION"])
neighborhoodWM = list(df["INCIDENTNEIGHBORHOOD"])
zoneWM = list(df["INCIDENTZONE"])
latWM = list(df["X"])
longWM = list(df["Y"])
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

trace1 = go.Bar(
    x= neighborhoodW,
    y= ageW,
    name='Women',
    text= offensesM,
    marker = dict( color='#F5B041')
)
trace2 = go.Bar(
    x= neighborhoodM,
    y= ageM,
    name='Men',
    text= offensesM,
    marker = dict( color = '#BFC9CA')
)

trace3 = go.Bar(
    x= neighborhoodWM,
    y= ageWM,
    name='White Men',
    text= offensesM,
    marker = dict( color = '#BFC9CA')
)
trace4 = go.Bar(
    x = neighborhoodWW,
    y= ageWW,
    name='White Women',
    text= offensesM,
    marker = dict( color ='#F5B041')
)

data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    #barmode='stack',
    title='Arrests in Allegheny County',
    updatemenus=list([
        dict(
            x=-0.05,
            y=1,
            yanchor='top',
            buttons=list([
                dict(
                    args=['visible', [True, True, False, False]],
                    label='Men and Women',
                    method='restyle'
                ),
                dict(
                    args=['visible', [False, False, True, True]],
                    label='White Women vs White Men',
                    method='restyle'
                ),
                dict(
                    args=['visible', [False, True, False, False]],
                    label='Silver',
                    method='restyle'
                ),
                dict(
                    args=['visible', [False, False, True, False]],
                    label='Bronze',
                    method='restyle'
                ),
                dict(
                    args=['visible', [False, False, False, True]],
                    label='Total',
                    method='restyle'
                )
            ]),
        )
    ]),
)


fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='stacked-bar')
```
